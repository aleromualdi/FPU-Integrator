```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fput.integrator import FPUT_Integrator
```

```python
fpu = FPUT_Integrator(
    num_atoms=32,
    num_modes=6,
    initial_mode_number=1,
    initial_mode_amplitude=1,
    t_step=0.05,
    t_max=200,
    alpha=0.0,
    beta=0.3
)

data = fpu.run(method="verlet")
```

```python
def plot_mode_energies(data, xlim=None):
    time_steps, _, _, mode_energies = data

    plt.plot(time_steps, mode_energies[0, :], linestyle='dashed', color='k', linewidth=1.5, label='Mode 1', alpha=0.5)
    plt.plot(time_steps, mode_energies[1, :], linestyle='dashed', color='b', linewidth=1.5, label='Mode 2', alpha=0.5)
    plt.plot(time_steps, mode_energies[2, :], linestyle='dashed', color='r', linewidth=1.5, label='Mode 3', alpha=0.5)
    plt.plot(time_steps, mode_energies[3, :], linestyle='dashed', color='y', linewidth=1.5, label='Mode 4', alpha=0.5)
    plt.plot(time_steps, mode_energies[4, :], linestyle='dashed', color='g', linewidth=1.5, label='Mode 5', alpha=0.5)
    if xlim:
        plt.xlim(xlim)
    plt.xlabel('$t$ ')
    plt.ylabel('Energy ')
    legend = plt.legend(loc='upper right', shadow=True, fontsize='x-small')
    plt.show()
```

```python
plot_mode_energies(data, xlim=(0, 1))
```
